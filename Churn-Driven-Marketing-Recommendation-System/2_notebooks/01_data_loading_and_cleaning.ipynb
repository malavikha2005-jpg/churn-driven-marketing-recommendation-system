```python
import pandas as pd
import numpy as np

```

```python
# Path to raw dataset
raw_path = "../_data/raw/online_retail_II.csv"

# Load dataset
df = pd.read_csv(raw_path)

# Check data
print(df.shape)
df.head()
```

```python
df = df[df['Customer ID'].notna()]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.columns
```

```python
df = df[df['Customer_ID'].notna()]
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df = df[df['Quantity'] > 0]
```

```python
df = df[df['Price'] > 0]
```

```python
df = df[df['Price'] > 0]
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['TotalAmount'] = df['Quantity'] * df['Price']
```

```python
print(df.shape)
df.describe()
```

```python
df.to_csv("../_data/processed/cleaned_transactions.csv", index=False)
```
