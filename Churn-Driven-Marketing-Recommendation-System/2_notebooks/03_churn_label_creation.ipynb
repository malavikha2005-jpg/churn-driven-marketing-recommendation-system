```python
import pandas as pd
```

```python
customer_df = pd.read_csv(
    "../_data/processed/customer_level_features.csv",
    parse_dates=["first_purchase", "last_purchase"]
)

customer_df.head()
```

```python
CHURN_THRESHOLD_DAYS = 180
```

```python
customer_df["churn"] = customer_df["recency_days"].apply(
    lambda x: 1 if x > CHURN_THRESHOLD_DAYS else 0
)
```

```python
customer_df["churn"].value_counts(normalize=True)
```

```python
customer_df.to_csv(
    "../_data/processed/modeling_dataset.csv",
    index=False
)

customer_df.head()
```
