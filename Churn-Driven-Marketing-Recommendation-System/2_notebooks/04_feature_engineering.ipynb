```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../_data/processed/modeling_dataset.csv")
df.head()
```

```python
df["total_revenue_log"] = np.log1p(df["total_revenue"])
df["avg_order_value_log"] = np.log1p(df["avg_order_value"])
df["total_quantity_log"] = np.log1p(df["total_quantity"])
df["total_transactions_log"] = np.log1p(df["total_transactions"])
```

```python
df.columns
```

```python
df = df.drop(
    columns=[
        "total_revenue",
        "avg_order_value",
        "total_quantity",
        "total_transactions"
    ]
)
```

```python
df["customer_lifetime_days_log"] = np.log1p(df["customer_lifetime_days"])
df = df.drop(columns=["customer_lifetime_days"])
```

```python
df = df.drop(columns=["Customer_ID"])
```

```python
df.info()
df.head()
```

```python
df.to_csv(
    "../_data/processed/final_features.csv",
    index=False
)
```
